```python
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
```

```python
img =  np.round(io.imread('http://mrl.cs.vsb.cz/people/gaura/dzo/uneq.jpg', as_gray=True)*255)
```

```python
img.shape # un canal
```

```python
img.max()
```

```python
img
```

```python
# graficar imagen
plt.figure(figsize=(14,14))
plt.imshow(img,cmap='gray')
```

```python
# histograma
plt.figure(figsize=(14,14))
plt.hist(img.ravel(),bins=range(0,256))
```

```python
# ecualizar imagen
nk , k = np.histogram(img.ravel(),bins=range(0,256))
# probabilidad de k
M = img.shape[0]
N = img.shape[1]
pk = nk/(M*N)
T1 = pk.cumsum()
kl = np.round(T1*255)
```

```python
np.round(kl)
```

```python
#nuevo histograma
plt.figure(figsize=(14,14))
plt.hist(img.ravel(),bins=range(0,256))
```

```python
img_eq = img.copy()
for i in k:
  if i == k[-1]:
    break
  img_eq[img==1] = kl[i]
  
```

```python
plt.figure(figsize=(14,14))
plt.imshow(img_eq,cmap='gray')
```

```python
plt.figure(figsize=(14,14))
plt.hist(img.ravel(),bins=range(0,256))
```

```python
fig, axs = plt.subplots(2,2, figsize=(20,20))
axs[0,0].imshow(img,cmap='gray')
axs[0,1].imshow(255-img_eq,cmap='gray')
axs[1,0].hist(img.ravel(), bins=range(0,256))
axs[1,1].hist(img_eq.ravel(), bins=range(0,256))
```

```python
#convolucion
Gx = np.array([[1,0,-1],
              [2,0,-2],
              [1,0,-1]])
imgconvx = np.zeros((img.shape[0]-Gx.shape[0]+1,img.shape[1]-Gx.shape[1]+1))
for i in range(0,img.shape[0]-Gx.shape[0]+1):
  for j in range(0,img.shape[1]-Gx.shape[1]+1):
    imgconvx[i,j] = np.sum(img[i:i+Gx.shape[0],j:j+Gx.shape[1]]*Gx)
```

```python
plt.figure(figsize=(14,14))
plt.imshow(imgconvx,cmap='gray')
```
