```python
# histograma de forma manual
plt.figure()
plt.hist(img.ravel(),bins=range(0,256));
```

```python
# mejorar la imagen
plt.imshow(equalize_hist(img),cmap='gray')
```

```python
# mejorar la imagen
img_eq = equalize_hist(img)
plt.hist(img_eq.ravel()*255,bins=range(0,256));
```

```python
# mejorar la imagen
img_gamma = adjust_gamma(img)
fig, ax = plt.subplots(1,2)
ax[0].hist(img_gamma.ravel()*255,bins=range(0,256));
ax[1].imshow(img_gamma,cmap='gray')
```

```python
# procesamiento morfologico de la imagen
from skimage.morphology import (dilation,erosion, closing, opening, disk, square)
```

```python
img2 = io.imread('https://raw.githubusercontent.com/jorgevergara/PDI/main/soap_image.jpg',as_gray = True)
```

```python
plt.figure()
plt.imshow(img2,cmap='gray')
```

```python
im_th = img2>0.7
plt.imshow(im_th,cmap='gray')
# 1 blanco / 0 negro
```

```python
# borrar los puntos de la imagen (erosion)
# primero erosion y luego dilatacion = opening
im_p1 = opening(im_th,disk(3))
plt.figure(figsize=(12,12))
plt.imshow(im_p1,cmap='gray')
```

```python
from skimage.filters import (gaussian,sobel,median,inverse) 
# gaussian para bordes
```

```python
im3 = io.imread('https://raw.githubusercontent.com/jorgevergara/PDI/main/building_image.jpg',as_gray=True)
plt.figure(figsize=(15,15))
plt.imshow(im3,cmap='gray')
```

```python
# suavizar imagen con filtro de promedio
im_md = median(im3,disk(3))
plt.figure(figsize=(15,15))
plt.imshow(im_md,cmap='gray')
```

```python
im_gs = gaussian(im3)
plt.figure(figsize=(15,15))
plt.imshow(im_gs,cmap='gray')
```

```python
im_sb = sobel(im3)
plt.figure(figsize=(15,15))
plt.imshow(im_sb,cmap='gray')
```
