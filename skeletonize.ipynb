```python
import numpy as np
import matplotlib.pylab as plt
import skimage.io as io
```

```python
from skimage.morphology import skeletonize
```

```python
img = io.imread('https://raw.githubusercontent.com/jorgevergara/PDI/main/horse-dog.jpg',as_gray=True)
```

```python
plt.figure()
plt.imshow(img,cmap='gray')
```

```python
img.max()
```

```python
im_binarizada = img>0.8
```

```python
ims = skeletonize(im_binarizada)
```

```python
plt.figure()
plt.imshow(ims.astype(int),cmap='gray')
```

```python
from skimage import data
```

```python
caballo = data.horse()
```

```python
plt.figure(figsize=(15,15))
plt.imshow(caballo,cmap="gray")
```

```python
ims2 = skeletonize(1-caballo)
ims2.max()
```

```python
plt.figure(figsize=(15,15))
plt.imshow(skeletonize(1-caballo),cmap="gray")
```
