```python
import skimage.io as io
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = io.imread('https://raw.githubusercontent.com/jorgevergara/PDI/main/parrot_thumb.jpg')
```

```python
plt.figure()
plt.imshow(img)
```

```python
img.shape
```

```python
img
```

```python
img.max()
```

```python
img.min()
```

```python
# bins = cuantas divisiones quiero hacer en mis datos (1,2,3)
# range = el rango en el que estamos trabajando ej: (0,255)
# color = color en el histograma
# alpha = transparencia del histograma
plt.figure()
plt.hist(img[:,:,0].ravel(),bins=256,range=(0,255),color='r',alpha=0.3)
plt.hist(img[:,:,1].ravel(),bins=256,range=(0,255),color='g',alpha=0.3)
plt.hist(img[:,:,2].ravel(),bins=256,range=(0,255),color='b',alpha=0.3)
# ravel() sirve para convertirse a un vector (array)
```

```python
im2 = img.copy()
gamma = 0.5
im2 = im2.astype(float) ** gamma
```

```python
plt.figure()
plt.imshow(im2/im2.max())
```

```python
im2.shape
```

```python
print(im2.min())
print(im2.max())
```

```python
# bins = cuantas divisiones quiero hacer en mis datos (1,2,3)
# range = el rango en el que estamos trabajando ej: (0,255)
# color = color en el histograma
# alpha = transparencia del histograma
plt.figure()
plt.hist(im2[:,:,0].ravel(),bins=256,range=(0,255),color='r',alpha=0.3)
plt.hist(im2[:,:,1].ravel(),bins=256,range=(0,255),color='g',alpha=0.3)
plt.hist(im2[:,:,2].ravel(),bins=256,range=(0,255),color='b',alpha=0.3)
# ravel() sirve para convertirse a un vector (array)
```

```python
from skimage import exposure
```

```python
img_eq = exposure.equalize_hist(img)
```

```python
plt.figure()
plt.imshow(img_eq)
```

```python
# bins = cuantas divisiones quiero hacer en mis datos (1,2,3)
# range = el rango en el que estamos trabajando ej: (0,255)
# color = color en el histograma
# alpha = transparencia del histograma
plt.figure()
plt.hist(img_eq[:,:,0].ravel(),bins=256,range=(0,1),color='r',alpha=0.3)
plt.hist(img_eq[:,:,1].ravel(),bins=256,range=(0,1),color='g',alpha=0.3)
plt.hist(img_eq[:,:,2].ravel(),bins=256,range=(0,1),color='b',alpha=0.3)
# ravel() sirve para convertirse a un vector (array)
```
